```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combine_df = pd.merge(study_results, mouse_metadata, how='outer', on="Mouse ID")

# XXXcombine_df = pd.concat([ study_results, mouse_metadata], ignore_index=True, sort=False)
# Display the data table for preview

combine_df

```

```python
# Checking the number of mice.
number_mice = combine_df["Mouse ID"].drop_duplicates()
number_mice = len(number_mice)
number_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
dup_mouseID_time = combine_df.loc[combine_df.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
dup_mouseID_time 
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_mouseID_time_df = combine_df.loc[combine_df["Mouse ID"] == "g989", :]
dup_mouseID_time_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

clean_mouse_df = combine_df.loc[combine_df["Mouse ID"]!="g989",:]
clean_mouse_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mice_num = clean_mouse_df["Mouse ID"].nunique()
clean_mice_num
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen


# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
drug_regimen_mean = clean_mouse_df['Tumor Volume (mm3)'].groupby(clean_mouse_df['Drug Regimen']).mean()
drug_regimen_median = clean_mouse_df['Tumor Volume (mm3)'].groupby(clean_mouse_df['Drug Regimen']).median()
drug_regimen_var = clean_mouse_df['Tumor Volume (mm3)'].groupby(clean_mouse_df['Drug Regimen']).var()
drug_regimen_std_deciation = clean_mouse_df['Tumor Volume (mm3)'].groupby(clean_mouse_df['Drug Regimen']).std()
drug_regimen_sem = clean_mouse_df['Tumor Volume (mm3)'].groupby(clean_mouse_df['Drug Regimen']).sem()


# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
mean_med_var_stand_sem_summary_df = pd.DataFrame({"Mean Tumor Volume":drug_regimen_mean, 
                            "Median Tumor Volume":drug_regimen_median, 
                           "Tumor Volume Variance":drug_regimen_var, 
                           "Tumor Volume Std. Dev.":drug_regimen_std_deciation, 
                           "Tumor Volume Std. Err.":drug_regimen_sem})


mean_med_var_stand_sem_summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
summary_Statistics_aggregation =  clean_mouse_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].aggregate(['mean', 'median', 'var', 'std', 'sem'])
summary_Statistics_aggregation
```

## Quartiles, Outliers and Boxplots

```python
####For review
clean_mouse_df
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse.
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint.

MT_df = clean_mouse_df[["Mouse ID","Timepoint"]].groupby("Mouse ID").max().reset_index()
merge_df = MT_df.merge(clean_mouse_df, how="left", on=["Mouse ID", "Timepoint"])
merge_df
```

```python
# Idea for easy pick up in ***Put treatments into a list for for loop (and later for plot labels)***
DrugRegimen_name_df = merge_df.set_index("Drug Regimen")
DrugRegimen_name_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
print(f"Capomulin, Ramicane, Infubinol, Ceftamin")
# Create empty list to fill with tumor vol data (for plotting)
tumorVol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for regimen in treatment_list:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes.
    subset = merge_df.loc[merge_df["Drug Regimen"] == regimen, 'Tumor Volume (mm3)']

    # add subset 
    tumorVol.append(subset)  
    
    # Determine outliers using upper and lower bounds
    quartiles = subset.quantile([.25,.5,.75])
    quartiles

    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    iqr = upperq-lowerq
    
    print("---------------------------------------------------------------------------")

    print(f"The lower quartile of treatment is: {lowerq}")
    print(f"The upper quartile of treatment is: {upperq}")
    print(f"The interquartile range of treatment is: {iqr}")
    print(f"The the median of treatment is: {quartiles[0.5]} ")
                          
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
       
    print(f"Values below {lower_bound} could be treatment outliers.")
    print(f"Values above {upper_bound} could be treatment outliers.")  
   
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
labels = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

fig1, ax1 = plt.subplots()
ax1.set_title('Tumor volume for each treatment group',fontsize =20)
ax1.set_ylabel('Tumor Volume (mm3)')
ax1.set_xlabel('Drug Regimen')
ax1.boxplot(tumorVol, labels=["Capomulin","Ramicane","Infubinol","Ceftamin",])

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# Part 1

avgCapomulin_vol =Capomulin_df.groupby(['Mouse ID']).mean()
avgCapomulin_vol
```

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
# Part 2


correlation_coefficient = round(st.pearsonr(avgCapomulin_vol['Weight (g)'],avgCapomulin_vol['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between between both factors is {correlation_coefficient}")

```

```python
# linear regression model Part 1
x_values = avgCapomulin_vol["Weight (g)"]
y_values = avgCapomulin_vol["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept

print(f"slope:{slope}")
print(f"intercept:{intercept}")
print(f"rvalue (Correlation coefficient):{rvalue}")
print(f"pandas (Correlation coefficient):{correlation_coefficient}")
print(f"stderr:{stderr}")

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)
```

```python
# linear regression model Part 2
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.title("Mouse Weight vs Tumor Volume for Capomulin")
plt.show()
```
